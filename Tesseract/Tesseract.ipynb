# Draw boxes

```python
box1 = pytesseract.image_to_boxes(img)
box2 = pytesseract.image_to_boxes(img2)
box3 = pytesseract.image_to_boxes(img3)
```

```python
print(box1)
```

```python
imgH, imgW, _ = img.shape
```

```python
for a in box1.splitlines():
    a = a.split(' ')
    x,y = int(a[1]), int(a[2])
    w,h = int(a[3]), int(a[4])
    cv2.rectangle(img,(x,imgH-y),(w,imgH-h),(0,0,255),1)
```

```python
plt.imshow(img)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```
